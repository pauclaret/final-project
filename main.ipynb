```python
import numpy as np
```

```python
df_numericas = merged_df.select_dtypes(include=np.number)
```

```python
df_numericas.columns
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
import pandas as pd

# Select only the numerical columns from the original dataframe
numerical_cols = merged_df.select_dtypes(include=[np.number]).columns

imputer = IterativeImputer()
imputer.fit(df_numericas[numerical_cols])
df_numericas_imputed = imputer.transform(df_numericas[numerical_cols])

numericas_trans = pd.DataFrame(df_numericas_imputed, columns=numerical_cols)
numericas_trans.head()
```

```python
columnas = numericas_trans.columns
df_cleaned = merged_df.drop(columnas, axis = 1)
```

```python
df_cleaned[columnas] = numericas_trans
df_cleaned.head()
```

```python
df_cleaned['wakeupTime'] = df_cleaned['wakeupTime'].apply(lambda x: round(float(x), 2))
```

```python
df_cleaned['wakeupTime'] = df_cleaned['wakeupTime'] * 0.01
```

```python
df_cleaned['wakeupTime'] = df_cleaned['wakeupTime'].round(2)
```

```python
df_cleaned.head()
```

```python
df_cleaned["Country or region"].unique
```

```python
df_cleaned.isnull().sum()
```

```python
df_cleaned["Value"] = df_cleaned["Value"].fillna("No")
```

```python
df_cleaned.columns
```

```python
df_cleaned["Deaths_Drugs"] = df_cleaned["Deaths - Amphetamine use disorders - Sex: Both - Age: All Ages (Number)"] + df_cleaned["Deaths - Cocaine use disorders - Sex: Both - Age: All Ages (Number)"] + df_cleaned["Deaths - Opioid use disorders - Sex: Both - Age: All Ages (Number)"]
```

```python
df_cleaned.loc[df_cleaned['Country or region'] == 'Central African Republic', 'Country or region'] = 'CAR'
```
